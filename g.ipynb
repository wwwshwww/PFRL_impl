```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [10, 6]
# plt.rcParams['figure.dpi'] = 100
```

```python
np.log(np.exp(1))
```

```python
disc = 200
x_max = 1
start_discount = 5

x = np.linspace(0, x_max, disc)
y = np.array(list(map(lambda x: x_max-np.log(x), x)))
plt.plot(x,y)
```

```python
e = np.linspace(np.exp(4), np.exp(1), disc//start_discount)
y2 = np.array([np.log(e[0])-(np.log(e[0])-np.log(n)) for n in e])
y2 -= np.min(y2)
y2 /= np.max(y2) - np.min(y2)
plt.plot(np.arange(disc//start_discount), y2)
```

```python
tab = np.linspace(np.exp(2), np.exp(1), disc//start_discount)
y3 = [np.log(tab[0])-np.log(n) for n in tab]
y[disc-disc//start_discount:] -= y2[::-1]
plt.plot(x, y)
```
